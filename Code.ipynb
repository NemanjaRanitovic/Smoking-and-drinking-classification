```python
import pandas as pd


# Convert all columns to string type
df = df.astype(str)

# Replace any problematic values with NaN
df = df.apply(pd.to_numeric, errors='coerce')

# Now, plot histograms for each column
for col in df.columns:
    plt.figure(figsize=(8, 6))
    for class_label in df['SMK_stat_type_cd'].unique():
        subset = df[df['SMK_stat_type_cd'] == class_label]
        plt.hist(subset[col], alpha=0.5, label=f'Class {class_label}')
    plt.title(f'Distribution of {col} by SMK_stat_type_cd')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```
